```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
file_path = "/Users/vinay/Downloads/diabetes.csv"
data = pd.read_csv(file_path)

# Display first few rows
data.head()
```

```python
# Assuming 'Outcome' is the target variable
X = data.drop('Outcome', axis=1)  # Independent variables (features)
y = data['Outcome']               # Dependent variable (target)

# Check feature names and target
print("Features:", X.columns.tolist())
print("Target variable:", y.name)
```

```python
# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

```python
# Build Decision Tree using entropy
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X_train, y_train)

# Check training accuracy
train_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=X.columns, class_names=['No Diabetes','Diabetes'], filled=True)
plt.show()
```

```python
# Functions to calculate metrics
def entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p*np.log2(p) - (1-p)*np.log2(1-p)

# Root node probability
p_root = y_train.mean()
entropy_root = entropy(p_root)
gini_root = 2 * p_root * (1 - p_root)

print("Entropy of root node:", entropy_root)
print("Gini Index of root node:", gini_root)

# Example: Information Gain for splitting on 'Glucose'
threshold = X_train['Glucose'].median()
left = y_train[X_train['Glucose'] <= threshold]
right = y_train[X_train['Glucose'] > threshold]

p_left = left.mean()
p_right = right.mean()

entropy_left = entropy(p_left)
entropy_right = entropy(p_right)

weight_left = len(left)/len(y_train)
weight_right = len(right)/len(y_train)

info_gain = entropy_root - (weight_left*entropy_left + weight_right*entropy_right)
print("Information Gain by splitting on Glucose:", info_gain)
```
